# Réaliser une ACP avec sklearn

```python
import pandas
import numpy
from sklearn.preprocessing import StandardScaler
```

```python
## Définition d'une matrice :
x = [ [12, 30, 80, -100],
     [-1000, 12, -23, 10],
     [14, 1000, 0, 0]
    ]

## Ensuite on peut obtenir un df ou une matrice np :
x_np = numpy.asarray(x)

x_pd = pandas.DataFrame(x)
```

```python
## Afficher les statistiques descriptives de la matrice :
x_pd.describe()
```

```python
## On instancie notre scaler (il faut normaliser les données pour les rendre comparable)
scaler = StandardScaler()

scaler.fit(x_pd) ## fitting

x_scaled = scaler.transform(x_pd) ## entraînement

## Puis on le transforme en df :
df_res = pandas.DataFrame(x_scaled)

## On affiche les statistiques descriptives de la matrice standardisée :

df_res.describe().round(2)
```
