```python
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
visites = numpy.array([7, 5, 35, 5, 6, 8, 9, 45, 8, 9, 10, 11, 42, 9, 11])
visites
```

```python
fig, ax = plt.subplots()

ax.plot(visites)
ax.set_title("Evolution du nombre d'entrées dans le musée sur 3 semaine")

plt.show()
```

Il y a visiblement une saisonnalité de période 5. Nous allons donc opérer un lissage par moyenne mobile sur 5 jours :

```python
musees = pandas.DataFrame(data={"x":visites, "jour":3 * [ "Lundi", "Mardi", "Mercredi", "Jeudi", "Vendredi" ]})
musees
```

```python
def GetMoment(p) :
    m = (p-1) / 2
    
    return int(m)

m = GetMoment(5) ## Attendu : 2

musees["M5"] = numpy.nan

## Calcul de la moyenne mobile ;
for i in range(m, musees.shape[0]-m) :
    min_index = i - m
    max_index = i + m
    
    musees.loc[i, "M5"] = musees.loc[min_index:max_index, "x"].mean() ## prendre entre le min et le max défini plus tôt.
    
musees
```

```python
## Calcul des coefficients de saisonnalités :
musees["diff"] = musees["x"] - musees["M5"]
musees

coeff_saison = musees.groupby(by=["jour"], as_index=False)["diff"].mean()
coeff_saison["theta"] = coeff_saison["diff"] - coeff_saison["diff"].mean(axis=0)
coeff_saison
```
